$$
 f_n = \left\{ 
 \begin{array}{lll}
 0& if & n=0\\
 1& if & n=1\\
 f_{n-1} +   f_{n-2} & if & n> 1 \text{ (otherwise)}\\
 \end{array}
   \right.
$$

```python
import itertools as tools
import matplotlib.pyplot as plt
from time import time

# Fibonacci Recursive 
def fRecursive(n):
  if n <=1:
    return n
  else:
    return (fRecursive(n-1)+fRecursive(n-2))  

# Fibonacci Iterartive 
def fIterative(n):
  a=0
  b=1
  for i in range(n):
    a=b
    b=a+b
  return a

# Fibonacci Logaritmitc Iterative
def fLog(n):
  if n==0: return 0 
  v1, v2, v3 = 1, 1, 0  
  for rec in bin(n)[3:]:
    calc = (v2*v2) 
    v1, v2, v3 = (v1*v1+calc) , ((v1+v3)*v2) , (calc+v3*v3) 
    if rec == '1': v1, v2, v3 = (v1+v2), v1, v2
  return v2;    
```

# PLOTS

```python
#Recursive Execution
RT = []
Rn = []
for i in range(50):
  Rn.append(i)
  start = time()
  fRecursive(i)
  finish = time()
  exectime = finish-start
  RT.append(exectime)
plt.title("FIBONACCI RECURSIVO")
plt.plot(Rn,RT,label="Recursivo")
plt.plot(Rn,Rn,label="Lineal")
plt.ylabel("Tiempo")
plt.legend()
plt.show()
```

```python
#Iterative
IT = []
In = []
for i in range(1000):
  In.append(i)
  start = time()
  fIterative(i)
  finish = time()
  exectime = finish-start
  IT.append(exectime)
plt.title("FIBONACCI ITERATIVO")
plt.plot(In,IT,label="Iterativo")
plt.plot(In,In,label="lineal")
plt.ylabel("Tiempo")
plt.legend()
plt.show()
```

```python
#Iterative Logarithmic Execution
IlogT = []
Ilogn = []
for i in range(1000):
  Ilogn.append(i)
  start = time()
  fLog(i)
  finish = time()
  exectime = finish-start
  IlogT.append(exectime)
plt.title("FIBONACCI ITERATIVO LOGARITMICO")
plt.plot(Ilogn,IlogT,label="Iterativo logaritmico")
plt.plot(Ilogn,Ilogn,label="Lineal")
plt.ylabel("Tiempo")
plt.legend()
plt.show()
```

# COMPARISON TIME EXECUTION


```python
IT = []
In = []
IlogT = []
Ilogn = []

for i in range(50):

  In.append(i)
  start = time()
  fIterative(i)
  finish = time()
  exectime = finish-start
  IT.append(exectime)

  Ilogn.append(i)
  start_ = time()
  fLog(i)
  finish_ = time()
  exectime_ = finish_-start_
  IlogT.append(exectime_)


plt.plot(Rn, RT, label='Recursivo')
plt.plot(Rn, IT, label='Iterativo')
plt.plot(Rn, IlogT, label='Iterativo logaritmico')
plt.plot(Rn,Rn)

plt.xlabel('n')
plt.ylabel('Tiempo')
plt.title("Tiempo de ejecucion")
plt.legend()
plt.show()
```
